```python
import pandas as pd
from fuzzywuzzy import fuzz
```

```python
# Import data
df = pd.read_csv("/Users/pol/Downloads/conflicts_to_match.csv", 
                 delimiter=';')
```

```python
df["similarity_score"] = df.apply(
    lambda row: fuzz.ratio(
        str(row["name_old_id"]).lower(),
        str(row["name_new_id"]).lower()
    ),
    axis=1
)
```

```python
# remove NAs
def safe_str(x):
    if x is None or pd.isna(x):
        return ""
    return str(x)

df["name_old_id"] = df["name_old_id"].apply(safe_str)
df["name_new_id"] = df["name_new_id"].apply(safe_str)
```

```python
df["score_ratio"] = df.apply(
    lambda row: fuzz.ratio(row["name_old_id"], 
                           row["name_new_id"]),
    axis=1
)

df["score_partial"] = df.apply(
    lambda row: fuzz.partial_ratio(row["name_old_id"], 
                                   row["name_new_id"]),
    axis=1
)

df["score_token_set"] = df.apply(
    lambda row: fuzz.partial_token_set_ratio(row["name_old_id"],
                                             row["name_new_id"]),
    axis=1
)
```

```python
df["combined_score"] = (
    0.2 * df["score_ratio"] +
    0.4 * df["score_partial"] +
    0.4 * df["score_token_set"]
)

df["match"] = ""
```

```python
df.to_excel("/Users/pol/Downloads/to_match_scores.xlsx", index=False)

print("Saved with similarity scores.")
```
